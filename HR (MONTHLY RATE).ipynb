# ANOVA(cat ~ con)

```python
def ANOVA(df,cat,con):
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    rel = con + "~" + cat
    model = ols(rel,df).fit()
    Q = anova_lm(model)
    pval = round(Q.iloc[0:1,4][0],4)
    return pval
```

```python
imp_cols = []
for i in cat:
    p = ANOVA(df,i,"MonthlyRate")
    print("--------MonthlyRate vs",i,"pval---------->",p)
    
    #if (p<0.05):
        #imp_cols.append(i)
```

# Define X and Y

```python
X = df.drop(labels= "MonthlyRate",axis=1)
Y = df[["MonthlyRate"]] 
```

# Preprocessing

```python
cat = []
con = []

for i in X.columns:
    if (X[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
        
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
X1 =pd.DataFrame(mm.fit_transform(X[con]),columns=con)
X2 = pd.get_dummies(X[cat])
X = X1.join(X2)
```

# Splitting data in training and testing

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(X , Y , test_size= 0.2 , random_state=21)
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(xtrain , ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
from sklearn.metrics import mean_absolute_error
tr_err = round(mean_absolute_error(ytrain,tr_pred),2)
ts_err = round(mean_absolute_error(ytest,ts_pred),2)
tr_err,ts_err,round(tr_err-ts_err)
```

# Model Object Function

```python
def modeller(mo):
    model = mo.fit(xtrain , ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err = round(mean_absolute_error(ytrain,tr_pred),2)
    ts_err = round(mean_absolute_error(ytest,ts_pred),2)
    return tr_err,ts_err,round(tr_err-ts_err)
```

# Different Algos for Regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
```

# Decision Tree

```python
dtr = DecisionTreeRegressor(random_state=21)
modeller(dtr)
```

# Pruning

```python
# max Depth
for i in range(2,20):
    dtr = DecisionTreeRegressor(random_state=21,max_depth=i)
    print(i , modeller(dtr))
```

```python
#min_samples_leaf
for i in range(2,20):
    dtr = DecisionTreeRegressor(random_state=21,min_samples_leaf=i)
    print(i , modeller(dtr))
```

```python
# min_samples_split
for i in range(2,20):
    dtr = DecisionTreeRegressor(random_state=21,min_samples_split=i)
    print(i , modeller(dtr))
```

# Random Forest

```python
for i in range(2,20):
    rfr = RandomForestRegressor(n_estimators= i,random_state=21)
    print(i , modeller(rfr))   
```

# Adaboost


```python
for i in range(2,20):
    abc = AdaBoostRegressor(DecisionTreeRegressor(random_state=21,max_depth=2),n_estimators=i)
    print(i , modeller(abc))  
```

# KNN

```python
for i in range(2,20):
    knc = KNeighborsRegressor(n_neighbors=i)
    print(i,modeller(knc))
```

* After trying different models we have got good results with Adaboost Regressor with n_estimators = 4
